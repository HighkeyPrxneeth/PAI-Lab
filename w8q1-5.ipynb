## 1. Rule-Based Logical Agent for Smart Lighting

A smart building uses sensors to control lighting automatically. The system must decide whether to turn lights **ON** or **OFF** based on logical rules.

### System Percepts

The system perceives:
- **MotionDetected**
- **IsDark**
- **ManualOverride**

### Knowledge Representation (Propositional Logic)

Represent the following knowledge:

- If it is dark **and** motion is detected, turn the light **ON**
- If there is **no motion**, turn the light **OFF**
- If **manual override is ON**, ignore all automatic rules

### Logical Agent Design

Design a logical agent that:

- Takes sensor readings as **percepts**
- Updates its **knowledge base**
- Infers the **correct action**

### Task

Given a set of percepts, determine the agent's action using **logical inference**.

```python
class SmartLightAgent:
    def __init__(self):
        self.light_on = False  # Initial state

    def decide(self, motion_detected, is_dark, manual_override):
        # Rule 1: Manual override has highest priority
        if manual_override:
            return "Manual Override Active → No automatic action"

        # Rule 2: If dark AND motion detected → Light ON
        if is_dark and motion_detected:
            self.light_on = True
            return "Light ON"

        # Rule 3: If no motion → Light OFF
        if not motion_detected:
            self.light_on = False
            return "Light OFF"

        # Default case
        return "No Change"


if __name__ == "__main__":
    agent = SmartLightAgent()

    # Test cases
    test_inputs = [
        (True, True, False),   # Motion + Dark
        (False, True, False),  # No motion
        (True, False, False),  # Motion but not dark
        (True, True, True)     # Manual override
    ]

    for i, (m, d, o) in enumerate(test_inputs, 1):
        result = agent.decide(m, d, o)
        print(f"Test Case {i}: Motion={m}, Dark={d}, Override={o} → {result}")
```

### Logical Design

If it is dark and motion is detected, turn light ON

`(D ∧ M) → L`

If there is no motion, turn light OFF

`!M → !L`

## 2. Medical Diagnosis Logical Agent

A medical decision-support system assists doctors by reasoning over symptoms to suggest possible conditions.

### Symptoms

Given symptoms:
- **Fever**
- **Cough**
- **ChestPain**

### Diseases

- **Flu**
- **Pneumonia**

### Knowledge Representation (Propositional Logic)

Represent the following medical knowledge:

- **Fever ∧ Cough → Flu**
- **Fever ∧ Cough ∧ ChestPain → Pneumonia**
- **Pneumonia → HospitalizationRequired**

### Task

Build a knowledge base and use **forward chaining** to infer:

- Possible **diseases**
- Required **actions**

### Conceptual Question

Explain how this agent differs from a **machine-learning classifier**.

```python
class MedicalAgent:
    def __init__(self):
        self.facts = set()

    def add_fact(self, fact):
        self.facts.add(fact)

    def forward_chain(self):
        inferred = True

        while inferred:
            inferred = False

            # Rule 1: Fever ∧ Cough → Flu
            if "Fever" in self.facts and "Cough" in self.facts:
                if "Flu" not in self.facts:
                    self.facts.add("Flu")
                    inferred = True

            # Rule 2: Fever ∧ Cough ∧ ChestPain → Pneumonia
            if ("Fever" in self.facts and 
                "Cough" in self.facts and 
                "ChestPain" in self.facts):
                if "Pneumonia" not in self.facts:
                    self.facts.add("Pneumonia")
                    inferred = True

            # Rule 3: Pneumonia → HospitalizationRequired
            if "Pneumonia" in self.facts:
                if "HospitalizationRequired" not in self.facts:
                    self.facts.add("HospitalizationRequired")
                    inferred = True

        return self.facts


if __name__ == "__main__":
    agent = MedicalAgent()

    # Input symptoms
    symptoms = ["Fever", "Cough", "ChestPain"]

    for symptom in symptoms:
        agent.add_fact(symptom)

    result = agent.forward_chain()

    print("Inferred Results:")
    for fact in result:
        print("-", fact)
```

### Logical Design 

`F ∧ C → Flu`

`F ∧ C ∧ P → Pneumonia`

`Pneumonia → Hospitalization Required`

## 3. Logical Agent for Traffic Violation Detection

A smart traffic monitoring system detects violations using logical rules rather than image-based ML models.

### Percepts

The system perceives:
- **SpeedAboveLimit**
- **SignalRed**
- **VehicleCrossedLine**
- **EmergencyVehicle**

### Knowledge Representation (Logical Rules)

Define logical rules to infer:

- **SpeedViolation**
- **SignalViolation**
- **Exceptions for emergency vehicles**

### Logical Agent Cycle

Implement the following agent cycle:

- **Perceive**
- **Update knowledge base**
- **Infer violations**
- **Decide enforcement action**

### Task

Given a sequence of percepts over time, show how conclusions change.

```python
class TrafficAgent:
    def __init__(self):
        self.facts = set()

    def perceive(self, percepts):
        self.facts = set(percepts)

    def infer(self):
        violations = []

        # Emergency vehicle → no violations
        if "EmergencyVehicle" in self.facts:
            return ["No Violation (Emergency Vehicle)"]

        # Rule 1: Speed violation
        if "SpeedAboveLimit" in self.facts:
            violations.append("SpeedViolation")

        # Rule 2: Signal violation
        if ("SignalRed" in self.facts and 
            "VehicleCrossedLine" in self.facts):
            violations.append("SignalViolation")

        if not violations:
            return ["No Violation"]

        return violations

    def act(self, violations):
        actions = []

        for v in violations:
            if v == "SpeedViolation":
                actions.append("Issue Speed Ticket")
            elif v == "SignalViolation":
                actions.append("Issue Signal Violation Ticket")

        if "No Violation (Emergency Vehicle)" in violations:
            actions.append("Allow Passage")

        if "No Violation" in violations:
            actions.append("No Action")

        return actions


# Simulation Over Time
if __name__ == "__main__":
    agent = TrafficAgent()

    # Sequence of percepts over time
    timeline = [
        ["SpeedAboveLimit"],  # t1
        ["SignalRed", "VehicleCrossedLine"],  # t2
        ["SpeedAboveLimit", "EmergencyVehicle"],  # t3
        ["SignalRed"],  # t4
        []  # t5
    ]

    for t, percepts in enumerate(timeline, 1):
        print(f"\nTime t{t}")
        print("Percepts:", percepts)

        agent.perceive(percepts)
        violations = agent.infer()
        actions = agent.act(violations)

        print("Violations:", violations)
        print("Action:", actions)
```

### Logical Design

1. Speed violation:

`S ∧ !E → SV`

2. Signal violation:

`R ∧ C ∧ !E → RV`

3. Emergency exception:

`E → !SV ∧ !RV`

## 4. Logical Agent for Cybersecurity Intrusion Detection

An enterprise network uses a logical agent to detect potential cyber intrusions based on system events.

### Percepts

The system perceives:
- **MultipleFailedLogins**
- **LoginFromUnknownIP**
- **AccessToSensitiveFiles**
- **AdminPrivileges**

### Knowledge Representation (Propositional Logic)

Encode the following security knowledge:

- **MultipleFailedLogins ∧ LoginFromUnknownIP → SuspiciousLogin**
- **SuspiciousLogin ∧ AdminPrivileges → HighRiskIntrusion**
- **HighRiskIntrusion → TriggerAlert**

### Logical Agent Design

Design a knowledge-based agent that:

- Continuously updates **facts**
- Infers **threat levels**
- Decides **defensive actions**

```python
class CyberSecurityAgent:
    def __init__(self):
        self.facts = set()

    def perceive(self, percepts):
        self.facts = set(percepts)

    def infer(self):
        inferred = True

        while inferred:
            inferred = False

            # Rule 1: MultipleFailedLogins ∧ LoginFromUnknownIP → SuspiciousLogin
            if ("MultipleFailedLogins" in self.facts and 
                "LoginFromUnknownIP" in self.facts):
                if "SuspiciousLogin" not in self.facts:
                    self.facts.add("SuspiciousLogin")
                    inferred = True

            # Rule 2: SuspiciousLogin ∧ AdminPrivileges → HighRiskIntrusion
            if ("SuspiciousLogin" in self.facts and 
                "AdminPrivileges" in self.facts):
                if "HighRiskIntrusion" not in self.facts:
                    self.facts.add("HighRiskIntrusion")
                    inferred = True

            # Rule 3: HighRiskIntrusion → TriggerAlert
            if "HighRiskIntrusion" in self.facts:
                if "TriggerAlert" not in self.facts:
                    self.facts.add("TriggerAlert")
                    inferred = True

        return self.facts

    def act(self):
        actions = []

        if "TriggerAlert" in self.facts:
            actions.append("Raise Security Alert")
            actions.append("Lock Account")
            actions.append("Notify Admin")

        elif "SuspiciousLogin" in self.facts:
            actions.append("⚠Flag Login Attempt")
            actions.append("Request Multi-Factor Authentication")

        else:
            actions.append("Normal Activity")

        return actions


# Simulation Over Time
if __name__ == "__main__":
    agent = CyberSecurityAgent()

    timeline = [
        ["MultipleFailedLogins"],  # t1
        ["MultipleFailedLogins", "LoginFromUnknownIP"],  # t2
        ["MultipleFailedLogins", "LoginFromUnknownIP", "AdminPrivileges"],  # t3
        ["AccessToSensitiveFiles"],  # t4
    ]

    for t, percepts in enumerate(timeline, 1):
        print(f"\nTime t{t}")
        print("Events:", percepts)

        agent.perceive(percepts)
        facts = agent.infer()
        actions = agent.act()

        print("Inferred Facts:", facts)
        print("Actions:", actions)
```

### Logical Design

1. Suspicious login:

`M ∧ U → SL`

2. High-risk intrusion:

`SL ∧ A → HR`

3. Trigger alert:

`HR → TA`

## 5. Autonomous Warehouse Robot Using Logical Reasoning

An autonomous warehouse robot must decide how to move and operate safely using logical reasoning rather than learning.

### Percepts

The robot perceives:
- **ObstacleAhead**
- **PathClear**
- **CarryingFragileItem**
- **HumanNearby**
- **DestinationReached**

### Knowledge Representation (Propositional Logic)

Represent the following operational rules:

- **ObstacleAhead → Stop**
- **CarryingFragileItem ∧ ObstacleAhead → SlowDown**
- **HumanNearby → ReduceSpeed**
- **PathClear ∧ ¬HumanNearby → MoveForward**

### Logical Agent Architecture

Design a logical agent architecture showing:

- **Knowledge base**
- **Inference engine**
- **Action selection**

```python
class WarehouseRobot:
    def __init__(self):
        self.facts = set()

    def perceive(self, percepts):
        self.facts = set(percepts)

    def infer_actions(self):
        actions = set()

        # Rule 1: ObstacleAhead → Stop
        if "ObstacleAhead" in self.facts:
            actions.add("Stop")

        # Rule 2: CarryingFragileItem ∧ ObstacleAhead → SlowDown
        if ("CarryingFragileItem" in self.facts and 
            "ObstacleAhead" in self.facts):
            actions.add("SlowDown")

        # Rule 3: HumanNearby → ReduceSpeed
        if "HumanNearby" in self.facts:
            actions.add("ReduceSpeed")

        # Rule 4: PathClear ∧ ¬HumanNearby → MoveForward
        if ("PathClear" in self.facts and 
            "HumanNearby" not in self.facts):
            actions.add("MoveForward")

        return actions

    def decide(self, actions):
        priority = ["Stop", "SlowDown", "ReduceSpeed", "MoveForward"]

        for action in priority:
            if action in actions:
                return action

        return "No Action"


# Simulation
if __name__ == "__main__":
    robot = WarehouseRobot()

    timeline = [
        ["PathClear"],  # t1
        ["ObstacleAhead"],  # t2
        ["ObstacleAhead", "CarryingFragileItem"],  # t3
        ["PathClear", "HumanNearby"],  # t4
        ["PathClear"],  # t5
    ]

    for t, percepts in enumerate(timeline, 1):
        print(f"\nTime t{t}")
        print("Percepts:", percepts)

        robot.perceive(percepts)
        actions = robot.infer_actions()
        decision = robot.decide(actions)

        print("Possible Actions:", actions)
        print("Final Decision:", decision)
```

### Logical Design

1. Obstacle handling:

`O → Stop`

2. Fragile item safety:

`F ∧ O → Slow Down`

3. Human safety:

`H → Reduce Speed`

4. Normal movement:

`P ∧ !H → Move Forward`
