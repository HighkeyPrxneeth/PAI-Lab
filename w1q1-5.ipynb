## 1. Implementation of a queue in Python using two stacks

**Description**: A queue can be implemented using two stacks in Python by following the below steps:
1. Use two stacks, stack1 and stack2, to implement the enqueue and dequeue operations.
2. In the enqueue operation, push the new element onto stack1.
3. In the dequeue operation, if stack2 is empty, transfer all elements from stack1 to stack2. The element at the top of stack2 is the first element that was pushed onto stack1 and thus represents the front of the queue. Pop this element from stack2 to return it as the dequeued element.

```python
class Queue:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []
    def enqueue(self, element):
        self.stack1.append(element)
        return True
    def dequeue(self):
        if len(self.stack2) == 0:
            while len(self.stack1) > 0:
                self.stack2.append(self.stack1.pop())
        return self.stack2.pop()
    def show(self):
        print("Queue: ", end="")
        for element in reversed(self.stack2):
            print(element, end=" ")
        for element in self.stack1:
            print(element, end=" ")
        print()
```

```python
queue = Queue()
queue.enqueue(69)
queue.enqueue(67)
queue.enqueue(420)
print("Enqueued 69, 67, 420")
queue.show()
d = queue.dequeue()
print(f"Dequeued {d}")
queue.show()
queue.enqueue(69)
print("Enqueued 69")
queue.show()
```

## 2. Implement the following graph using python. Print the adjacency list and adjacency matrix.

[A graph is a data structure that consists of vertices that are connected via edges.]

![image](graph.png)

```python
class Edge:
    def __init__(self, src, dst, weight):
        self.src = src
        self.dst = dst
        self.weight = weight

class Node:
    def __init__(self, label, connections):
        self.label = label
        self.edges = []
        for connection in connections:
            self.edges.append(Edge(self.label, connection[0], connection[1]))

class Graph:
    def __init__(self, nodes):
        self.nodes = nodes
        self.adj_list = self._generate_adj_list()
        self.adj_mat = self._generate_adj_mat()
    def _generate_adj_list(self):
        adj_list = {}
        for node in self.nodes:
            adj = []
            for edge in node.edges:
                adj.append([edge.dst, edge.weight])
            adj_list[node.label] = adj
        return adj_list
    def _generate_adj_mat(self):
        mat = []
        size = len(self.nodes)
        for node in self.nodes:
            node_list = [0] * size
            for edge in node.edges:
                node_list[edge.dst - 1] = edge.weight
            mat.append(node_list)
        return mat
    def display(self):
        print(self.adj_list)
        print(self.adj_mat)
```

```python
node1 = Node(1, [(2, 1), (3, 1)])
node2 = Node(2, [(3, 3)])
node3 = Node(3, [(4, 4)])
node4 = Node(4, [(1, 5)])
nodes = [node1, node2, node3, node4]
graph = Graph(nodes)
graph.display()
```

## 3. Create two list X and Y with some set of numerical values. Compute Euclidean distance for corresponding values in X and Y. Store the distance values in a separate list and sort them using Bubble sort algorithm.

![Formula](formula.png)

```python
from math import sqrt

X = [5, 31, 51, 80, 20, 36, 67, 85, 82, 20]
Y = [70, 30, 92, 87, 1, 65, 59, 59, 9, 58]
dist = []

def compute_distance(x, y):
    return sqrt((y - x) ** 2)

for x, y in zip(X, Y):
    dist.append(compute_distance(x, y))

def bubble_sort(arr):
    for i in range(len(arr)):
        highest = 0
        highest_idx = 0
        for j in range(len(arr)-i):
            if arr[j] > highest:
                highest = arr[j]
                highest_idx = j
        temp = arr[len(arr)-i-1]
        arr[len(arr)-i-1] = arr[highest_idx]
        arr[highest_idx] = temp
    return arr

print("Sorted distances: ")
print(bubble_sort(dist))
```

## 4. Implement the following directed unweighted graph using class, methods, and data structures of Python.

![Unweighted Graph](unweighted_graph.png)

**Expected output**:<br>
(0 —> 1)<br>
(1 —> 2)<br>
(2 —> 0) (2 —> 1)<br>
(3 —> 2)<br>
(4 —> 5)<br>
(5 —> 4)

```python
class Edge:
    def __init__(self, src, dst):
        self.src = src
        self.dst = dst

class Node:
    def __init__(self, label, connections):
        self.label = label
        self.edges = []
        for connection in connections:
            self.edges.append(Edge(self.label, connection))

class Graph:
    def __init__(self, nodes):
        self.nodes = nodes
    def display(self):
        for node in self.nodes:
            for edge in node.edges:
                print(f"({edge.src} -> {edge.dst})", end=" ")
            print()
```

```python
node0 = Node(0, [1])
node1 = Node(1, [2])
node2 = Node(2, [0, 1])
node3 = Node(3, [2])
node4 = Node(4, [5])
node5 = Node(5, [4])

graph = Graph([node0, node1, node2, node3, node4, node5])
graph.display()
```

## 5. Implement the following undirected weighted graph using class, methods, and data structures of Python. Print the adjacency list and adjacency matrix.

![Undirected Graph](undirected_graph.png)

Expected Output:<br>
Adjacency List:<br>
["A:['B',<br>
'C', 'E']",<br>
"C:['A',<br>
'B', 'D',<br>
'E']",<br>
"B:['A',<br>
'C', 'D']",<br>
"E:['A',<br>
'C']",<br>
"D:['B',<br>
'C']"]<br>
Adjacency Matrix<br>
&nbsp;&nbsp;&nbsp;&nbsp;[[ 0. 1. 1. 0. 1.]<br>
&nbsp;&nbsp;&nbsp;&nbsp;[ 1. 0. 1. 1. 0.]<br>
&nbsp;&nbsp;&nbsp;&nbsp;[ 1. 1. 0. 1. 1.]<br>
&nbsp;&nbsp;&nbsp;&nbsp;[ 0. 1. 1. 0. 0.]<br>
&nbsp;&nbsp;&nbsp;&nbsp;[ 1. 0. 1. 0. 0.]]

```python
class Node:
    def __init__(self, label, connections):
        self.label = label
        self.connections = connections
        
class Graph:
    def __init__(self, nodes):
        self.nodes = nodes
        self.adj_mat = self._generate_adj_mat()
    def _generate_adj_mat(self):
        mat = []
        size = len(self.nodes)
        for node in self.nodes:
            con_set = set(node.connections)
            node_list = [0] * size
            for i in range(size):
                if chr(65+i) in con_set:
                    node_list[i] = 1
            mat.append(node_list)
        return mat
    def display(self):
        for node in self.nodes:
            print(f"{node.label}: ", end=" ")
            for connection in node.connections:
                print(connection, end=" ")
            print()
        print(self.adj_mat)
```

```python
a = Node("A", ["B", "C", "E"])
b = Node("B", ["A", "C"])
c = Node("C", ["A", "B", "D", "E"])
d = Node("D", ["C"])
e = Node("E", ["A", "C"])

graph = Graph([a, b, c, d, e])
graph.display()
```
