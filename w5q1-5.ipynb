### Experiment 2: A* Search for Optimal Maze Navigation

**Problem Statement:**

Consider a maze represented as a graph, where each node corresponds to a location in the maze and edges represent valid paths between locations.

An agent is placed at a start node and must reach a goal node.

To ensure optimal navigation, the agent will use the A* search algorithm, which combines:

**Actual cost from start 𝑔(𝑛) and Heuristic estimate to goal ℎ(𝑛)**

**Task:**
1. Formulate the maze navigation problem as a search problem by clearly defining: State space, Initial state, Actions, Transition model, Goal test, Path cost, Heuristic function ℎ(𝑛)
2. Implement the A* Search algorithm to find the optimal shortest path from the start node to the goal node.
3. Use the evaluation function: `𝑓(𝑛)=𝑔(𝑛)+ℎ(𝑛)`
   
    where:
   
        𝑔(𝑛) = path cost from start to node 𝑛
   
        ℎ(𝑛) = Manhattan distance from node 𝑛 to goal
   <br>
5. Print the path returned by A* Search.
6. Count and display the number of nodes expanded during the search.

**Constraints:**

* Each move has uniform cost = 1.
* The heuristic must be admissible (never overestimates).
* The environment is fully observable, deterministic, static, discrete, and single-agent.
* A* must be implemented without using any built-in graph search libraries.
* The algorithm should always expand the node with the lowest 𝑓(𝑛)

```python
# A* Search for Maze Navigation

def manhattan_distance(node, goal):
    return abs(node[0] - goal[0]) + abs(node[1] - goal[1])


def a_star_search(maze, start, goal):
    rows, cols = len(maze), len(maze[0])

    open_list = [(start, 0, manhattan_distance(start, goal))]  # (node, g, h)
    visited = set()
    parent = {}
    g_cost = {start: 0}

    nodes_expanded = 0

    while open_list:
        # Select node with minimum f(n) = g(n) + h(n)
        open_list.sort(key=lambda x: x[1] + x[2])
        current, g, h = open_list.pop(0)

        nodes_expanded += 1

        if current == goal:
            # Reconstruct path
            path = []
            while current in parent:
                path.append(current)
                current = parent[current]
            path.append(start)
            path.reverse()
            return path, nodes_expanded

        visited.add(current)

        x, y = current
        neighbors = [(x-1, y), (x+1, y), (x, y-1), (x, y+1)]

        for nx, ny in neighbors:
            if 0 <= nx < rows and 0 <= ny < cols:
                if maze[nx][ny] == 0:
                    neighbor = (nx, ny)
                    new_g = g + 1

                    if neighbor not in g_cost or new_g < g_cost[neighbor]:
                        g_cost[neighbor] = new_g
                        h_value = manhattan_distance(neighbor, goal)
                        parent[neighbor] = current
                        if neighbor not in visited:
                            open_list.append((neighbor, new_g, h_value))

    return None, nodes_expanded


# Maze: 0 = free cell, 1 = wall
maze = [
    [0, 0, 0, 1],
    [1, 1, 0, 1],
    [0, 0, 0, 0],
    [0, 1, 1, 0]
]

start = (0, 0)
goal = (3, 3)

path, expanded = a_star_search(maze, start, goal)

print("Optimal path found:", path)
print("Number of nodes expanded:", expanded)
```

**BFS** explores the maze level by level and always finds the shortest path, but it expands many nodes and is inefficient for large mazes.

**Greedy Best-First Search** uses only a heuristic to move toward the goal, expanding fewer nodes and working faster, but it does not guarantee the shortest path.

**A*** combines actual path cost and heuristic information, expanding fewer nodes than BFS while still guaranteeing an optimal shortest path, making it the best balance between efficiency and optimality.

### Experiment 3: Recursive Best-First Search (RBFS) for Maze Navigation

**Problem Statement:**

Consider a maze represented as a graph, where each node corresponds to a location in the maze and edges represent valid paths between locations.  
An agent is placed at a start node and must reach a goal node by traversing the maze.

Traditional A* search requires storing all explored nodes in memory, which may become infeasible for large mazes.  
Recursive Best-First Search (RBFS) is a memory-bounded heuristic search algorithm that uses recursion and limited memory while still aiming to find optimal solutions.

**Task Requirements:**
1. Formulate the maze navigation problem as a search problem by clearly defining:
    * State space
    * Initial state
    * Actions
    * Transition model
    * Goal test
    * Path cost
    * Heuristic function \( h(n) \)
2. Implement the Recursive Best-First Search (RBFS) algorithm to find an optimal path from the start node to the goal node.
3. Use the evaluation function:
    * \( f(n) = g(n) + h(n) \)  
      where:
      * \( g(n) \) = path cost from the start node to node \( n \)
      * \( h(n) \) = Manhattan distance from node \( n \) to the goal
4. Print and display:
    * The path returned by RBFS
    * The total number of nodes expanded during the search

**Constraints:**
* The maze is unweighted (each step cost = 1).
* The environment is fully observable, deterministic, static, discrete, and single-agent.
* RBFS must be implemented without using built-in graph search libraries.
* The algorithm should use only linear memory through recursion.

**Expected Outcome:**

The algorithm should successfully find an optimal path from the start node to the goal while using limited memory, demonstrating the effectiveness of Recursive Best-First Search in memory-constrained environments.


```python
def manhattan(n, goal):
    return abs(n[0] - goal[0]) + abs(n[1] - goal[1])

nodes_expanded = 0

def rbfs(maze, node, goal, g, f_limit, path, visited):
    global nodes_expanded
    nodes_expanded += 1

    f = g + manhattan(node, goal)

    if node == goal:
        return path + [node], f

    successors = []
    x, y = node
    for nx, ny in [(x-1,y),(x+1,y),(x,y-1),(x,y+1)]:
        if 0 <= nx < len(maze) and 0 <= ny < len(maze[0]):
            if maze[nx][ny] == 0 and (nx, ny) not in visited:
                successors.append((nx, ny))

    if not successors:
        return None, float('inf')

    f_values = {}
    for s in successors:
        f_values[s] = max(g + 1 + manhattan(s, goal), f)

    while True:
        best = min(f_values, key=f_values.get)
        if f_values[best] > f_limit:
            return None, f_values[best]

        alternative = min(
            [f_values[s] for s in f_values if s != best],
            default=float('inf')
        )

        visited.add(best)
        result, new_f = rbfs(
            maze, best, goal, g + 1,
            min(f_limit, alternative),
            path + [node],
            visited
        )
        visited.remove(best)

        f_values[best] = new_f

        if result is not None:
            return result, new_f


# ---------- DRIVER CODE ----------

maze = [
    [0, 0, 0, 1],
    [1, 1, 0, 1],
    [0, 0, 0, 0],
    [0, 1, 1, 0]
]

start = (0, 0)
goal = (3, 3)

nodes_expanded = 0
path, cost = rbfs(
    maze,
    start,
    goal,
    g=0,
    f_limit=float('inf'),
    path=[],
    visited=set([start])
)

print("Path found:", path)
print("Nodes expanded:", nodes_expanded)
```

### Experiment 4: Iterative Deepening A* (IDA*) for Maze Navigation

**Problem Statement:**

Consider a maze represented as a graph, where each node corresponds to a location in the maze and edges represent valid paths between locations.  
An agent must reach the goal node efficiently. While A* search guarantees optimality, it requires large memory.

Iterative Deepening A* (IDA*) overcomes this limitation by combining depth-first traversal, the A* heuristic cost function, and iterative deepening on cost limits. IDA* is especially useful when memory is limited.

**Task Requirements:**
1. Formulate the maze navigation problem as a search problem by clearly defining:
    * State space
    * Initial state
    * Actions
    * Transition model
    * Goal test
    * Path cost
    * Heuristic function \( h(n) \)
2. Implement the Iterative Deepening A* (IDA*) algorithm to find the shortest path from the start node to the goal node.
3. Use the cost function:
    * \( f(n) = g(n) + h(n) \)
4. Perform iterative deepening by increasing the threshold value until a solution is found.
5. Print and display:
    * The path returned by IDA*
    * The total number of nodes expanded during the search

**Constraints:**
* The maze is unweighted (uniform step cost = 1).
* The heuristic function must be admissible.
* The environment is fully observable, deterministic, static, discrete, and single-agent.
* IDA* must be implemented without using built-in search libraries.
* The algorithm must use depth-first search with cost-bounded iterations.

**Expected Outcome:**

The algorithm should successfully find the shortest path from the start node to the goal while using significantly less memory than A*, demonstrating the efficiency of Iterative Deepening A* in memory-constrained environments.


```python
def ida_star(maze, start, goal):
    threshold = manhattan(start, goal)
    path = [start]
    expanded = 0

    def search(node, g, threshold):
        nonlocal expanded
        expanded += 1
        f = g + manhattan(node, goal)
        if f > threshold:
            return f
        if node == goal:
            return "FOUND"

        min_cost = float('inf')
        x, y = node
        for nx, ny in [(x-1,y),(x+1,y),(x,y-1),(x,y+1)]:
            if 0 <= nx < len(maze) and 0 <= ny < len(maze[0]):
                if maze[nx][ny] == 0 and (nx,ny) not in path:
                    path.append((nx,ny))
                    t = search((nx,ny), g+1, threshold)
                    if t == "FOUND":
                        return "FOUND"
                    min_cost = min(min_cost, t)
                    path.pop()
        return min_cost

    while True:
        temp = search(start, 0, threshold)
        if temp == "FOUND":
            return path, expanded
        threshold = temp
```

```python
start = (0, 0)
goal = (3, 3)

path, expanded = ida_star(maze, start, goal)

print("Optimal path found:", path)
print("Number of nodes expanded:", expanded)
```

### Experiment 5: Simplified Memory-Bounded A* (SMA*) for Maze Navigation

**Problem Statement:**

Consider a maze represented as a graph, where each node corresponds to a location in the maze and edges represent valid paths between locations.  
An agent must reach the goal node using an optimal strategy.

While A* search guarantees optimality, it may run out of memory when the search space becomes large.  
Simplified Memory-Bounded A* (SMA*) is a variant of A* that operates under a fixed memory limit by:
- Expanding the best nodes first
- Dropping the worst leaf nodes when memory is full
- Retaining optimality when sufficient memory is available

**Task Requirements:**
1. Formulate the maze navigation problem as a search problem by clearly defining:
    * State space
    * Initial state
    * Actions
    * Transition model
    * Goal test
    * Path cost
    * Heuristic function \( h(n) \)
2. Implement the Simplified Memory-Bounded A* (SMA*) algorithm to find the shortest path from the start node to the goal node.
3. Use the evaluation function:
    * \( f(n) = g(n) + h(n) \)
4. Introduce a fixed memory limit (maximum number of nodes stored).
5. When the memory limit is reached:
    * Remove the leaf node with the highest \( f(n) \) value
    * Backup its cost to its parent node
6. Print and display:
    * The final path returned by SMA*
    * The total number of nodes expanded during the search

**Constraints:**
* Uniform step cost = 1.
* The heuristic function must be admissible.
* The environment is fully observable, deterministic, static, discrete, and single-agent.
* SMA* must be implemented without using built-in graph search libraries.
* The algorithm must respect a predefined memory limit.

**Expected Outcome:**

The algorithm should successfully find the optimal path from the start node to the goal while operating within a fixed memory limit, demonstrating the effectiveness of Simplified Memory-Bounded A* in memory-constrained environments.


```python
def sma_star(maze, start, goal, memory_limit=10):
    open_list = [(start, 0, manhattan(start, goal))]
    parent = {}
    g_cost = {start: 0}
    expanded = 0

    while open_list:
        open_list.sort(key=lambda x: x[1] + x[2])

        # Enforce memory limit
        if len(open_list) > memory_limit:
            open_list.pop()   # remove worst f(n)

        node, g, h = open_list.pop(0)
        expanded += 1

        if node == goal:
            path = []
            while node in parent:
                path.append(node)
                node = parent[node]
            path.append(start)
            return path[::-1], expanded

        x, y = node
        for nx, ny in [(x-1,y),(x+1,y),(x,y-1),(x,y+1)]:
            if 0 <= nx < len(maze) and 0 <= ny < len(maze[0]):
                if maze[nx][ny] == 0:
                    neighbor = (nx, ny)
                    new_g = g + 1

                    if neighbor not in g_cost or new_g < g_cost[neighbor]:
                        g_cost[neighbor] = new_g
                        parent[neighbor] = node
                        open_list.append(
                            (neighbor, new_g, manhattan(neighbor, goal))
                        )

    return None, expanded
```

```python
start = (0, 0)
goal = (3, 3)

path, expanded = sma_star(maze, start, goal)

print("Optimal path found:", path)
print("Number of nodes expanded:", expanded)
```
