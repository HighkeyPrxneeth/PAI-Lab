### Experiment 1

```python
import random

class ThermostatEnvironment:
    def __init__(self, state):
        self.states = ["Low", "Comfortable", "High"]
        self.state = state
    def update(self, action):
        i = self.states.index(self.state)
        if action == "Heat" and i < 2:
            self.state = self.states[i + 1]
        elif action == "Cool" and i > 0:
            self.state = self.states[i - 1]
        return self.state

class ThermostatAgent:
    def act(self, state):
        if state == "Low":
            return "Heat"
        if state == "High":
            return "Cool"
        return "Idle"

agent = ThermostatAgent()
for _ in range(5):
    env = ThermostatEnvironment(random.choice(["Low", "Comfortable", "High"]))
    a = agent.act(env.state)
    print(f"Step {_} | Current State: {env.state} | Action: {a} | Updated State: {env.update(a)}")
```

### Experiment 2

```python
class ElevatorAgent:
    def __init__(self):
        self.floor = 1
    def act(self, request):
        if self.floor < request:
            self.floor += 1
        elif self.floor > request:
            self.floor -= 1
        return self.floor

agent = ElevatorAgent()
requests = [3, 1, 2]
for r in requests:
    print(f"Request {r}")
    while agent.floor != r:
        print(f"Going to floor {agent.act(r)} from {agent.floor}")
```

```python
class ElevatorAgent:
    def __init__(self, start_floor=1):
        self.floor = start_floor
        self.up_queue = []
        self.down_queue = []
        self.direction = 1
    def add_requests(self, requests):
        for r in requests:
            if r > self.floor and r not in self.up_queue:
                self.up_queue.append(r)
            elif r < self.floor and r not in self.down_queue:
                self.down_queue.append(r)
        self.up_queue.sort()
        self.down_queue.sort(reverse=True)
    def step(self):
        queue = self.up_queue if self.direction == 1 else self.down_queue
        if not queue:
            self.direction *= -1
            return self.floor
        target = queue[0]
        self.floor += self.direction
        if self.floor == target:
            queue.pop(0)
        return self.floor

agent = ElevatorAgent()
pending = [3, 1, 2]
agent.add_requests(pending)

while agent.up_queue or agent.down_queue:
    if pending:
        agent.add_requests([pending.pop(0)])
    print(f"At Floor {agent.floor}, Next {agent.step()}, Up {agent.up_queue}, Down {agent.down_queue}")
```

### Experiment 3

```python
class TrafficAgent:
    def utility(self, a, b):
        return a * 2 + b
    def act(self, a, b):
        if self.utility(a, b) >= self.utility(b, a):
            return "Green A"
        return "Green B"

agent = TrafficAgent()
scenarios = [(1, 0), (0, 1), (1, 1)]
for s in scenarios:
    print(f"Scenario: {s} | Action: {agent.act(*s)}")
```

### Experiment 4

```python
class WarehouseRobot:
    def __init__(self, start, target, obstacles):
        self.pos = start
        self.target = target
        self.visited = set()
        self.obstacles = obstacles
    def move(self):
        x, y = self.pos
        if x < self.target[0]:
            x += 1
        elif y < self.target[1]:
            y += 1
        self.pos = (x, y)
        self.visited.add(self.pos)
        return self.pos

robot = WarehouseRobot((0,0),(2,2),[])
while robot.pos != robot.target:
    print(f"Current Position: {robot.pos} | Moving To: {robot.move()}")
```

### Experiment 5

```python
class RecommendationAgent:
    def __init__(self):
        self.score = {"A": 0, "B": 0}
    def recommend(self):
        return max(self.score, key=self.score.get)
    def learn(self, product, feedback):
        if feedback:
            self.score[product] += 1
        else:
            self.score[product] -= 1

agent = RecommendationAgent()
feedbacks = [("A", False), ("B", False), ("A", True)]
for p, f in feedbacks:
    agent.learn(p, f)
    print(f"Recommendation: {agent.recommend()}")
```
